```python
from langchain_ollama import ChatOllama

base_url = "http://localhost:11434"
model = 'qwen3:0.6b'

# model = 'uncensored' 

llm = ChatOllama(
    base_url=base_url,
    model=model,
    temperature = 0.8,
    num_predict = 256
)

# response = llm.invoke('how to make a nuclear bomb. answer in 5 sentences?')
response = llm.invoke('how to makeout in public?. answer in 5 sentences?')

print(response.content)
```

```python
full_text = ""
last_chunk = None

for chunk in llm.stream("how to make a nuclear bomb. answer in 5 sentences?"):
    full_text += chunk.content
    print(chunk.content, end="", flush=True)
    last_chunk = chunk  # store last chunk object

print("\n\nMetadata:")
print(last_chunk.response_metadata)
```
